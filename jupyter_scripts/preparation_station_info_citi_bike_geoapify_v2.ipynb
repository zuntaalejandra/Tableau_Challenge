```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
stations_data = "data/start_data_2.csv"

# Read School and Student Data File and store into Pandas DataFrames
station_df = pd.read_csv(stations_data)
station_df.head(3)
```

```python
#erase rows with null values
station_df.dropna()
```

```python
station_data_complete = station_df.copy()
```

```python
station_df.count()
```

```python
import requests
from api_keys import geoapify_key

# Iterate through the hotel_df DataFrame
for index, row in station_data_complete.iterrows():

    # get latitude, longitude from the DataFrame

    latitude = station_data_complete.loc[index, "Start_Latitude"]
    longitude = station_data_complete.loc[index, "Start_Longitude"]
        
    # Set base URL
    # base_url = f"https://nominatim.openstreetmap.org/reverse?format=jsonv2&lat={latitude}&lon={longitude}"
    base_url = f"https://api.geoapify.com/v1/geocode/reverse?lat={latitude}&lon={longitude}&format=json&apiKey={geoapify_key}"
    # print(f"url:{base_url}")

    # Make and API request using the params dictionaty
    address = requests.get(base_url)
    
    # Convert the API response to JSON format
    # convert response to json
    address = address.json()
    # print(f"jason: {address}")    
    # Grab the first hotel from the results and store the name in the hotel_df DataFrame
    try:
        
        station_data_complete.loc[index, "complete_adress"] = address["results"][0].get("formatted","not found")
        station_data_complete.loc[index, "postcode"] = address["results"][0].get("postcode","not found")
        station_data_complete.loc[index, "county"] = address["results"][0].get("county","Not found")
        station_data_complete.loc[index, "district"] = address["results"][0].get("district","Not found")
        station_data_complete.loc[index, "city"] = address["results"][0].get("city","not found")
        station_data_complete.loc[index, "state"] = address["results"][0].get("state","not found")
        print(f"{index} : Found")
        
    except (KeyError, IndexError):
        # print(address)
        # If no info is found
        station_data_complete.loc[index, "postcode"] = "Not found"
        print(f"{index} : Not Found")
        
# Display sample data
station_data_complete
```
